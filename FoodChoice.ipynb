# How does income influence food choices?

## 1. Food Prices Analysis

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load food prices data
food_prices = pd.read_csv("C:\\Users\\Rodrigue Irampa\\Desktop\\Food Prices.csv")

# Display basic info
print("=== Food Prices Dataset ===")
print(f"Shape: {food_prices.shape}")
print("\nFirst 5 rows:")
display(food_prices.head())
```

### Summary statistics

```python
print("\nSummary Statistics:")
display(food_prices.describe(include='all'))
```

### Visualize monthly sales trends

```python
# Filter for sales data only
sales_data = food_prices[food_prices['Attribute'] == 'Total weighted monthly sales in U.S. dollars']

plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_data, x='Year', y='Value', hue='Metroregion_code')
plt.title('Monthly Food Sales Trends by Metro Region')
plt.ylabel('Sales (USD)')
plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Metro Region')
plt.tight_layout()
plt.show()
```

## 2. Dietary Habits Analysis

```python
# Load dietary habits data
dietary = pd.read_csv("C:\\Users\\Rodrigue Irampa\\Desktop\\Dietary Habits Survey Data.csv")

# Display basic info
print("\n=== Dietary Habits Dataset ===")
print(f"Shape: {dietary.shape}")
print("\nFirst 5 rows:")
display(dietary.head())

# First get the full column name for meals
meal_col = [col for col in dietary.columns if 'meals do you have' in col][0]
```

### Age and gender distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dietary, x='Age', hue='Gender')
plt.title('Age and Gender Distribution of Survey Respondents')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Gender')
plt.tight_layout()
plt.show()
```

### Diet type distribution

```python
plt.figure(figsize=(10, 6))
diet_counts = dietary['What would best describe your diet:'].value_counts()
diet_counts.plot(kind='bar')
plt.title('Distribution of Diet Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Meal frequency analysis

```python
print("\nMeal Frequency Statistics:")
print(dietary[meal_col].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(data=dietary, x=meal_col, order=dietary[meal_col].value_counts().index)
plt.title('Daily Meal Frequency Distribution')
plt.xlabel('Number of meals per day')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Income-Expenditure Analysis

```python
# Load income-expenditure data
income = pd.read_csv("C:\\Users\\Rodrigue Irampa\\Desktop\\Income-Expenditure.csv")

# Display basic info
print("\n=== Income-Expenditure Dataset ===")
print(f"Shape: {income.shape}")
print("\nFirst 5 rows:")
display(income.head())
```

### Income vs Expense scatter plot

```python
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(data=income, x='Mthly_HH_Income', y='Mthly_HH_Expense', 
                         hue='No_of_Fly_Members', size='No_of_Earning_Members',
                         sizes=(50, 200))
plt.title('Monthly Income vs Expenditure')
plt.xlabel('Monthly Household Income')
plt.ylabel('Monthly Household Expense')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Education level analysis

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=income, x='Highest_Qualified_Member', y='Annual_HH_Income')
plt.title('Annual Income by Education Level')
plt.xlabel('Highest Qualification')
plt.ylabel('Annual Household Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Cross-Dataset Insights (Without Merging)

```python
# Calculate key metrics from each dataset
print("\n=== Key Insights Across Datasets ===")

# From Food Prices
avg_sales = food_prices[food_prices['Attribute'] == 'Total weighted monthly sales in U.S. dollars']['Value'].mean()
print(f"\n1. Average monthly food sales: ${avg_sales:,.2f}")

# From Dietary Habits
veg_percentage = (dietary['What would best describe your diet:'].str.contains('Vegetarian').mean()) * 100
print(f"2. {veg_percentage:.1f}% of respondents follow some form of vegetarian diet")

# From Income-Expenditure
income_ratio = income['Mthly_HH_Expense'].sum() / income['Mthly_HH_Income'].sum()
print(f"3. Average expenditure to income ratio: {income_ratio:.2f} (spend {income_ratio*100:.0f}% of income)")
```

### Create a composite visualization

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Food sales distribution
sns.histplot(sales_data['Value'], bins=50, ax=axes[0], kde=True)
axes[0].set_title('Food Sales Distribution')
axes[0].set_xscale('log')
axes[0].set_xlabel('Sales Amount (USD)')

# Meal frequency
sns.countplot(data=dietary, x=meal_col, 
             order=dietary[meal_col].value_counts().index,
             ax=axes[1])
axes[1].set_title('Daily Meal Frequency')
axes[1].set_xlabel('Number of meals per day')
axes[1].tick_params(axis='x', rotation=45)

# Income distribution
sns.histplot(data=income, x='Annual_HH_Income', bins=20, ax=axes[2], kde=True)
axes[2].set_title('Annual Income Distribution')
axes[2].set_xlabel('Annual Household Income (USD)')

plt.tight_layout()
plt.show()
```

# Additional Focused Analysis

```python
# Focused analysis on dietary patterns by age
print("\n=== Dietary Patterns by Age Group ===")
# Dietary patterns by age
age_diet = dietary.groupby('Age')['What would best describe your diet:'].value_counts(normalize=True).unstack()

plt.figure(figsize=(12, 6))
age_diet.plot(kind='bar', stacked=True)
plt.title('Diet Type Distribution by Age Group')
plt.ylabel('Proportion')
plt.xlabel('Age Group')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Diet Type')
plt.tight_layout()
plt.show()
```

### Income by household size

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=income, x='No_of_Fly_Members', y='Annual_HH_Income')
plt.title('Annual Income by Household Size')
plt.xlabel('Number of Family Members')
plt.ylabel('Annual Household Income (USD)')
plt.tight_layout()
plt.show()
```
